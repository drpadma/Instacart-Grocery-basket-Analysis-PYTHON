```python
# Creating a Loyalty flag for the customers using transform
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

# Creating a loyalty flag using loc()

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge['loyalty_flag'].value_counts(dropna= False)
```

```python
ords_prods_merge.head(20)
```

#  Trying to determine is whether the prices of products purchased by loyal customers differ from those purchased by regular or new customers.

```python
ords_prods_merge.groupby('loyalty_flag').agg({'prices':['mean','min','max']})
```

```python
#Running descriptive statistics
ords_prods_merge['prices'].describe()
```

From seeing the descriptive statistics above its interesting to the average prices of new customers are higher when compared to the loyal customer.
The minimum and maximum remained same for all the three categories.

```python
#Creating a spending flag
ords_prods_merge['average_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.max)
```

```python
ords_prods_merge.head()
```

# Creating a spending flag

```python
ords_prods_merge.loc[ords_prods_merge['average_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['average_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
ords_prods_merge.head()
```

```python
#Checking for frequency
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

# Determining Frequent vs Non frequent customers

```python
#Creating a median flag using transform
ords_prods_merge['median_days_prior_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.max)
```

```python
ords_prods_merge.head()
```

# Creating a frequency flag

```python
ords_prods_merge.loc[ords_prods_merge['median_days_prior_order'] > 20, 'frequency_flag'] = 'Non frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_days_prior_order'] <= 20)& (ords_prods_merge['median_days_prior_order'] > 10),'frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days_prior_order'] <=10, 'frequency_flag'] = 'frequent customer'
```

```python
ords_prods_merge.head(20)
```

```python
#Checking for value points
ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```
